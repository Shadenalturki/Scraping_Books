```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
urls = []

first_url = 'https://books.toscrape.com/index.html'
urls.append(first_url)

for i in range(1, 51):
      urls.append(f'https://books.toscrape.com/catalogue/page-{i}.html')

print(urls)
```

```python
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
book_box = soup.find_all('article', class_='product_pod')
print(book_box)
```

```python
books_urls = []

for url in urls:
    for book in book_box:
        book_link = book.find('h3').find('a')['href']
        books_urls.append(book_link)

print(books_urls)
```

```python
books_names = []
books_ratings = []
books_prices = []
books_descriptions = []
books_upcs = []
books_availabilities = []

for url in books_urls:
    full_url = "https://books.toscrape.com/catalogue/" + url
    response = requests.get(full_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    title = soup.find('h1').text
    books_names.append(title)

    rating = soup.find('p', class_='star-rating')['class'][1]
    books_ratings.append(rating)

    price = soup.find('p', class_='price_color').text
    books_prices.append(price)

    description = soup.find('div', class_='sub-header').find_next('p').text
    books_descriptions.append(description)

    upc = soup.find('tr').find_next('td').text
    books_upcs.append(upc)

    availability = soup.find('p', class_='instock availability').text.strip()
    books_availabilities.append(availability)

for i in range(len(books_names)):
    print(f"Title: {books_names[i]}")
    print(f"Rating: {books_ratings[i]}")
    print(f"Price: {books_prices[i]}")
    print(f"Description: {books_descriptions[i]}")
    print(f"UPC: {books_upcs[i]}")
    print(f"Availability: {books_availabilities[i]}")
    print("\n")
```

```python
df = pd.DataFrame({
    'Book Name': books_names,
    'Rating': books_ratings,
    'Price': books_prices,
    'Description': books_descriptions,
    'UPC': books_upcs,
    'Availability': books_availabilities
})
df
```

```python
df.to_csv('Scraped_Books.csv')
```
